```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel(r"C:\Users\KRITHIKA\Downloads\Telco_customer_churn.xlsx")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Churn Reason'] = le.fit_transform(data['Churn Reason'])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data['Churn Reason'].nunique()
```

```python
data = pd.get_dummies(data, columns=['Contract', 'Paperless Billing', 'Payment Method'])
```

```python
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
data.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Gender', data=data, palette='Set2')
plt.title('Distribution of Customers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Contract Type Distribution
plt.figure(figsize=(8, 4))
sns.countplot(x='Contract_Month-to-month', data=data, palette='Set3')
sns.countplot(x='Contract_One year', data=data, palette='Set3')
sns.countplot(x='Contract_Two year', data=data, palette='Set3')
plt.title('Distribution of Customers by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Count')
plt.show()

# Paperless Billing Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Paperless Billing_Yes', data=data, palette='Set2')
plt.title('Distribution of Customers by Paperless Billing')
plt.xlabel('Paperless Billing')
plt.ylabel('Count')
plt.show()

# Payment Method Distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Payment Method_Bank transfer (automatic)', data=data, palette='Set3')
sns.countplot(x='Payment Method_Credit card (automatic)', data=data, palette='Set3')
sns.countplot(x='Payment Method_Electronic check', data=data, palette='Set3')
sns.countplot(x='Payment Method_Mailed check', data=data, palette='Set3')
plt.title('Distribution of Customers by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
numeric_columns = data.select_dtypes(include=['number']).columns
```

```python

correlation_matrix = data[numeric_columns].corr()

# 3. Plot the Heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
